```python
#my personal apriori algorithm

import numpy as np

#create subsets for pruning
def create_subsets(itemset, k):
    subsets = []

    #use bitmasking to save time and not have to use higher time complexity
    for bitmask in range(1 << len(itemset)):
        if bin(bitmask).count("1") == k:
            s = tuple(itemset[i] for i in range(len(itemset)) if bitmask & (1 << i))
            subsets.append(s)

    return subsets 

#self joining
def self_join(t1, t2):
    t1 = sorted(t1)
    t2 = sorted(t2)

    if t1[:-1] == t2[:-1]:
        return tuple(sorted(t1 + [t2[-1]]))

    return None #otherwise we can't merge so return none

#we are assuming that we are using a count support value in tbhis apriori implementation
def apriori_algorithm(df, minsup): 
    k = 0 #represents k = 1 (but we are using array indexing)

    temp = [tuple([col]) for col in df.columns]
    candidate_frequency = [{t : 0 for t in temp}, {}]
    l_frequent_itemsets = [[]]

    while len(candidate_frequency[k]) > 0:
        for key in candidate_frequency[k]:
            candidate_frequency[k][key] = np.sum(df[list(key)].all(axis=1))
        
        l_frequent_itemsets[k] = [itemset for itemset, frequency in candidate_frequency[k].items() if frequency > minsup]

        #generate candidates Ck+1

        candidates = set()
        #self-joining
        for i in range(len(l_frequent_itemsets[k])):
            for y in range(i+1, len(l_frequent_itemsets[k])):
                self_joined = self_join(l_frequent_itemsets[k][i], l_frequent_itemsets[k][y]) 
                if(self_joined is not None):
                        candidates.add(self_joined)

        #pruning
        if k > 2:
            for c in candidates:
                subsets = create_subsets(c, k)
                freq = True

                for subset in subsets:
                    if subset not in l_frequent_itemsets[k]:
                        freq = False
                        break
                if freq:
                    candidate_frequency[k+1][c] = 0

        k += 1

        candidate_frequency.append({})
        l_frequent_itemsets.append([])

    return l_frequent_itemsets[:-1]

a = apriori_algorithm(transactional_df, 0.02)
len(a)
```

```python
import numpy as np
import pandas as pd

#pivot table for organization. all values must be filled once we do this
item_m = training_df.pivot(index="user_id", columns="item_id", values="rating").fillna(0)

#i just used the algorithm in class here
U, SIG, Vt = np.linalg.svd(item_m, full_matrices=False)
SIG = np.diag(SIG)

def based_on_k(k):
    U2 = U[:, :k]
    SIG2 = SIG[:k, :k]
    Vt2 = Vt[:k, :]
    return np.dot(np.dot(U2, SIG2), Vt2)

#same for test dataset
test_m = testing_df.pivot(index="user_id", columns="item_id", values="rating").fillna(0)

#map over the index and columns for user/movie map, then perform rest of algorithm
user_map = {user: u for u, user in enumerate(item_m.index)}
movie_map = {movie: m for m, movie in enumerate(item_m.columns)}

test_users = [user_map[user] for user in test_m.index if user in user_map]
test_movies = [movie_map[movie] for movie in test_m.columns if movie in movie_map]

#labels must be flattened
true_labels = test_m.values.flatten() 
```

```python
#3b, 4a, and 4b
from sklearn.metrics import root_mean_squared_error, mean_absolute_error

#need to create boolean true or false masks to apply to the reconstructed matrix, so we can successfully get predictions
u_mask = np.isin(item_m.index, test_m.index)
movie_mask = np.isin(item_m.columns, test_m.columns)

ks = [5, 10, 20, 40]
for k in ks:
    reconstructed = based_on_k(k)
    #get predictions based on boolean masks & reconstruction matrix
    y_pred = reconstructed[u_mask][:, movie_mask].flatten() #3b, these are the pure predictions on unseen data
    
    print(f"K{k}: RMSE: {root_mean_squared_error(true_labels, y_pred)} and MAE: {mean_absolute_error(true_labels, y_pred)}")
```

The impact of k on model performance seems to be overall negative, with a higher k producing worse evaluation metrics. K controls the number of latent factors in the matrix. What seemed to have happened is that a higher k value captured more and more factors in our training data, which likely overfit on it, leading to worse performance when we tested on the testing data. The main strength of matrix factorization is that the results are interpretable and we can learn patterns that we otherwise wouldn't be able to find. The main downside as shown in b) is the fact that overfitting can occur. Another problem is that it relies on pre-existing data similar to content-based recommender systems, which leads to a cold start problem.
